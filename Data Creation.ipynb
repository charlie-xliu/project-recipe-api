```python
data = []
for i in range (4700 ,5000):
  response = requests.get(f'https://api.spoonacular.com/recipes/{i}/information?apiKey={api_key2}&includeNutrition=true')
  recipe = response.json()

  name = recipe.get('title')
  servings = recipe.get('servings')
  cookTime = recipe.get('readyInMinutes')
  link = recipe.get('sourceUrl')
  vegetarian = recipe.get('vegetarian')
  vegan = recipe.get('vegan')
  glutenFree = recipe.get('glutenFree')
  dairyFree = recipe.get('dairyFree')
  ingredients = recipe.get('extendedIngredients')
  cuisines = recipe.get('cuisines')
  dishTypes = recipe.get('dishTypes')
  diets = recipe.get('diets')
  score = recipe.get('spoonacularScore')

  d = {'name' : name,
       'servings' : servings,
       'cookTime' : cookTime,
       'link' : link,
       'vegetarian' : vegetarian,
       'vegan' : vegan,
       'glutenFree' : glutenFree,
       'dairyFree' : dairyFree,
       'ingredients' : ingredients,
       'cuisines' : cuisines,
       'dishTypes' : dishTypes,
       'diets' : diets,
       'score' : score}
  data.append(d)

df = pd.DataFrame(data)
df
```
